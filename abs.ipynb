```python
import torch 
import transformers
from torch import nn
class PegasusForMLM(nn.Module):
    """
    BertForMLM

    Simplified huggingface model
    """

    def __init__(
        self,
        model_name: str = "tuner007/pegasus_paraphrase",
        output_logits: bool = True,
        output_hidden_states: bool = True,
    ):
        """
        Args:
            model_name: huggingface model name
            output_logits: same as in huggingface
            output_hidden_states: same as in huggingface
        """
        super().__init__()
        self.config = transformers.AutoConfig.from_pretrained(
            model_name,
            output_hidden_states=output_hidden_states,
            output_logits=output_logits
        )
        self.bert = transformers.PegasusForCausalLM.from_pretrained(
            model_name, config=self.config
        )

    def forward(self, *model_args, **model_kwargs):
        """Forward method"""
        return self.bert(*model_args, **model_kwargs)
# from transformers import PegasusForMLM
model = torch.load('pg.pt')
```

```python
class MyLinearLayer(nn.Module):
    """ Custom Linear layer but mimics a standard linear layer """
    def __init__(self, size_in, size_out, weights, bias):
        super().__init__()
        self.size_in, self.size_out = size_in, size_out
        w = torch.Tensor(size_out, size_in)
        self.weights = nn.Parameter(w)  # nn.Parameter is a Tensor that's a module parameter.
        b = torch.Tensor(size_out)
        self.bias = nn.Parameter(b)
        self.weights = weights  # nn.Parameter is a Tensor that's a module parameter.
        self.bias = bias


    def forward(self, x):
        w_times_x= torch.mm(x, self.weights.t())
        return torch.add(w_times_x, self.bias)  # w times x + b
```
